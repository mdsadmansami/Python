```python
df = pd.DataFrame({
    'Cricket_Bat':['SG','BDM','SS','GM','KOOKABURRA','SPARTAN'],
    'MRP':[2000,2200,2400,2700,2800,3000],
    'Weight_Grams':[1100,1200,1250,1330,1480,1600]
})
```

```python
plt.plot(df['MRP'],df['Weight_Grams'])
plt.grid()
```

```python

plt.xlabel("Bat Price (USD)")
plt.ylabel("Bat Weight (grams)")
plt.plot(df['MRP'],df['Weight_Grams'])
plt.grid()
```

```python

plt.xlabel("Bat Price (USD)")
plt.ylabel("Bat Weight (grams)")
plt.plot(df['MRP'],df['Weight_Grams'])
plt.title('Sadman',loc='left')
plt.grid()
```

```python
a = np.arange(5)
b = [18,17,16,15,14]
c = [11,12,13,14,15]
```

```python
import matplotlib.pyplot as plt

ax=plt.subplot()
ax.plot(a,b,'k--',label='frequency')
ax.plot(a,c,'k:',label='Periods')
ax.legend()
```

```python
import matplotlib.pyplot as plt

ax=plt.subplot()
ax.plot(a,b,'k--',label='frequency')
ax.plot(a,c,'k:',label='Periods')
ax.legend(loc = 'lower center')
```

```python
import matplotlib.pyplot as plt

ax=plt.subplot()
ax.plot(a,b,'k--',label='frequency')
ax.plot(a,c,'k:',label='Periods')
l = ax.legend(loc = 'lower center')
l.get_frame().set_facecolor('green')
plt.xlabel('humy')
plt.ylabel('sami')
plt.title('learning matplotlib')
```

```python
import matplotlib.pyplot as plt

ax=plt.subplot()
ax.plot(a,b,'k--',label='frequency')
ax.plot(a,c,'k:',label='Periods')
l = ax.legend(loc = 'lower center',fontsize = 'x-large')
l.get_frame().set_facecolor('green')
plt.xlabel('humy')
plt.ylabel('sami')
plt.title('learning matplotlib')
```

```python
import matplotlib.pyplot as plt
x = pd.DataFrame({
    'Name':['Sadman','Humyra','Sami','Adiba'],
    'Grades':[3.75,4.00,3.50,3.75]
}) 
plt.bar(x['Name'],x['Grades'])
plt.xlabel('Name of the students')
plt.ylabel('Grades in love')
plt.title('Exam of LOVE')
```

```python
import matplotlib.pyplot as plt
x = pd.DataFrame({
    'Name':['Sadman','Humyra','Sami','Adiba'],
    'Scores':[10,50,5,35]
}) 
plt.pie(x['Scores'], labels = x['Name'], autopct = '%1f %%')
plt.title('A couple test of LOVE: participating HuMy & Sami')
```
